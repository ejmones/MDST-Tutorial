<h2> Feature Engineering </h2>

After exploring the data, you can choose to modify features that you would use to predict the performance of the students on your chosen response variable. 

You can also create your own features. For example, perhaps you figured that maybe a state's expenditure per student may affect their overall academic performance so you create a expenditure_per_student feature.

```python
#Variables of interest: [TOTAL_EXPENDITURE, INSTRUCTION_EXPENDITURE, SUPPORT_SERVICES_EXPENDITURE]


df["expenditure_per_student"] = df["INSTRUCTION_EXPENDITURE"] / df["ENROLL"]
df["expenditure_per_math_score"] = df["expenditure_per_student"] / df["AVG_MATH_8_SCORE"]
print(df)
```

<h2> Data Creation </h2>

_Use this space to create train/test data_

```python
from sklearn.model_selection import train_test_split
```

```python
#X <- input data
#Y <- response / output data

X = df[["YEAR","expenditure_per_student"]].dropna()
y = df.loc[X.index]["AVG_MATH_8_SCORE"] 

#fill  NaN with median of Y
y.fillna(y.median(), inplace = True)
```

```python
 X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# create your model here
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
model.fit(X_train,y_train)
```

```python
print(model.intercept_)
#["YEAR","expenditure_per_student"]
print(model.coef_)
```

```python
y_pred = model.predict(X_test)
model.score(X_test,y_test)
#print(np.mean(model.predict(X_test)-y_test))
```
